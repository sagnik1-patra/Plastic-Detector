```python
import os
import json
import pickle
from pathlib import Path
import itertools
import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.models import load_model

# =========================
# CONFIG (Windows paths)
# =========================
OUTPUT_DIR = r"C:\Users\sagni\Downloads\Plastic Detector"
DATA_ROOT  = r"C:\Users\sagni\Downloads\Plastic Detector\archive\dataset-resized"

MODEL_H5   = str(Path(OUTPUT_DIR) / "model.h5")
CLASS_PKL  = str(Path(OUTPUT_DIR) / "class_indices.pkl")
METRICS_JSON = str(Path(OUTPUT_DIR) / "metrics.json")

# Where to save new visuals/reports
ACC_PNG   = str(Path(OUTPUT_DIR) / "accuracy_loss_view.png")
CM_PNG    = str(Path(OUTPUT_DIR) / "confusion_matrix_view.png")
CR_CSV    = str(Path(OUTPUT_DIR) / "classification_report_view.csv")
CM_CSV    = str(Path(OUTPUT_DIR) / "confusion_matrix_view.csv")

# Must match training
IMG_SIZE   = (224, 224)
BATCH_SIZE = 32
VAL_SPLIT  = 0.2
SEED       = 42

# =========================
# Helpers
# =========================
def assert_exists(p, msg=None):
    if not Path(p).exists():
        raise FileNotFoundError(msg or f"Missing: {p}")

def load_class_indices(pkl_path):
    with open(pkl_path, "rb") as f:
        class_indices = pickle.load(f)
    # Ensure labels are in index order 0..C-1
    idx_to_class = {v: k for k, v in class_indices.items()}
    ordered_classes = [idx_to_class[i] for i in range(len(idx_to_class))]
    return class_indices, idx_to_class, ordered_classes

def build_val_generator(data_root, classes):
    datagen = ImageDataGenerator(
        preprocessing_function=preprocess_input,
        validation_split=VAL_SPLIT
    )
    val_gen = datagen.flow_from_directory(
        data_root,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        classes=classes,
        class_mode="categorical",
        shuffle=False,           # IMPORTANT: keep order intact for CM
        subset="validation",
        seed=SEED
    )
    return val_gen

def show_and_save_accuracy_loss(metrics_json_path, out_png):
    """
    Draw combined accuracy + loss curves from metrics.json (saved by training script).
    If missing, skip gracefully with a message.
    """
    if not Path(metrics_json_path).exists():
        print(f"[WARN] metrics.json not found at {metrics_json_path}; skipping accuracy/loss plot.")
        return

    with open(metrics_json_path, "r", encoding="utf-8") as f:
        m = json.load(f)

    hist = m.get("history", {})
    acc = hist.get("accuracy", [])
    val_acc = hist.get("val_accuracy", [])
    loss = hist.get("loss", [])
    val_loss = hist.get("val_loss", [])

    if not (acc and val_acc and loss and val_loss):
        print("[WARN] metrics.json exists but history arrays are empty; skipping accuracy/loss plot.")
        return

    epochs = range(1, len(acc) + 1)

    fig = plt.figure(figsize=(10, 8))

    ax1 = fig.add_subplot(2, 1, 1)
    ax1.plot(epochs, acc, label="Train Acc")
    ax1.plot(epochs, val_acc, label="Val Acc")
    ax1.set_title("Accuracy"); ax1.set_xlabel("Epoch"); ax1.set_ylabel("Accuracy")
    ax1.grid(alpha=0.25); ax1.legend(loc="lower right")

    ax2 = fig.add_subplot(2, 1, 2)
    ax2.plot(epochs, loss, label="Train Loss")
    ax2.plot(epochs, val_loss, label="Val Loss")
    ax2.set_title("Loss"); ax2.set_xlabel("Epoch"); ax2.set_ylabel("Loss")
    ax2.grid(alpha=0.25); ax2.legend(loc="upper right")

    fig.tight_layout()
    fig.savefig(out_png, dpi=200)
    print(f"[INFO] Saved accuracy/loss plot → {out_png}")
    plt.show()

def plot_confusion_matrix(cm, labels, out_png, normalize=True, title="Confusion Matrix (Normalized)"):
    """
    Heatmap with annotations (counts + % if normalized).
    """
    cm = np.array(cm)
    if normalize:
        cm_norm = cm.astype("float") / cm.sum(axis=1, keepdims=True)
        cm_norm = np.nan_to_num(cm_norm)
        display_mat = cm_norm
    else:
        display_mat = cm

    fig = plt.figure(figsize=(9, 7))
    ax = plt.gca()
    im = ax.imshow(display_mat, interpolation="nearest", cmap="viridis")
    plt.title(title)
    cbar = plt.colorbar(im, fraction=0.046, pad=0.04)
    cbar.ax.set_ylabel("Proportion" if normalize else "Count", rotation=90)
    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks, labels, rotation=45, ha="right")
    plt.yticks(tick_marks, labels)

    # annotate
    thresh = display_mat.max() / 2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if normalize:
            perc = display_mat[i, j] * 100.0
            txt = f"{cm[i, j]}\n{perc:.1f}%"
        else:
            txt = f"{cm[i, j]}"
        ax.text(j, i, txt,
                ha="center", va="center",
                color="white" if display_mat[i, j] > thresh else "black",
                fontsize=9)

    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.tight_layout()
    plt.savefig(out_png, dpi=220)
    print(f"[INFO] Saved confusion matrix heatmap → {out_png}")
    plt.show()

# =========================
# Main
# =========================
if __name__ == "__main__":
    # Basic checks
    assert_exists(MODEL_H5,   "Trained model not found. Train first to create model.h5.")
    assert_exists(CLASS_PKL,  "class_indices.pkl not found. Train first to create it.")
    assert_exists(DATA_ROOT,  "DATA_ROOT not found. Check your dataset path.")
    for sub in ["cardboard","glass","metal","paper","plastic","trash"]:
        assert_exists(Path(DATA_ROOT)/sub, f"Missing class folder: {sub}")

    # 1) Load model & class mapping
    print("[INFO] Loading model and class mapping…")
    model = load_model(MODEL_H5)
    class_indices, idx_to_class, ordered_classes = load_class_indices(CLASS_PKL)
    print("[INFO] Classes (ordered):", ordered_classes)

    # 2) Build validation generator (same split as training)
    print("[INFO] Building validation generator…")
    val_gen = build_val_generator(DATA_ROOT, ordered_classes)

    # 3) Predict on validation set
    print("[INFO] Predicting on validation set…")
    probs = model.predict(val_gen, verbose=1)
    y_pred = np.argmax(probs, axis=1)
    y_true = val_gen.classes  # indices aligned with 'ordered_classes'

    # 4) Classification report & confusion matrix
    labels_order = [idx_to_class[i] for i in range(len(ordered_classes))]
    report_dict = classification_report(
        y_true, y_pred, target_names=labels_order, output_dict=True, zero_division=0
    )
    # Save report CSV
    pd.DataFrame(report_dict).to_csv(CR_CSV)
    print(f"[INFO] Saved classification report → {CR_CSV}")

    cm = confusion_matrix(y_true, y_pred, labels=list(range(len(labels_order))))
    # Save raw counts CSV
    pd.DataFrame(cm, index=labels_order, columns=labels_order).to_csv(CM_CSV)
    print(f"[INFO] Saved confusion matrix (counts) → {CM_CSV}")

    # 5) Draw & show accuracy/loss curves (from metrics.json if available)
    show_and_save_accuracy_loss(METRICS_JSON, ACC_PNG)

    # 6) Draw & show confusion matrix (normalized)
    plot_confusion_matrix(
        cm, labels_order, CM_PNG, normalize=True,
        title="Confusion Matrix (Normalized)"
    )

    # 7) Quick, human-friendly summary on console
    acc = (y_pred == y_true).mean()
    print("\n================ SUMMARY ================")
    print(f"Timestamp: {datetime.datetime.now().isoformat(timespec='seconds')}")
    print(f"Validation images: {len(y_true)}")
    print(f"Overall Accuracy: {acc:.4f}")
    print("Per-class F1 (macro preview):")
    f1s = [report_dict.get(lbl, {}).get("f1-score", np.nan) for lbl in labels_order]
    for lbl, f1 in zip(labels_order, f1s):
        print(f"  - {lbl:>10s}: {f1:.3f}")
    print("=========================================\n")
```
