```python
import os
from pathlib import Path
from typing import List, Dict, Tuple, Optional

import numpy as np
import pandas as pd
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.efficientnet import preprocess_input

# ----------------------------
# CONFIG (Windows paths)
# ----------------------------
OUTPUT_DIR = Path(r"C:\Users\sagni\Downloads\Plastic Detector")
DATA_ROOT  = Path(r"C:\Users\sagni\Downloads\Plastic Detector\archive\dataset-resized")

MODEL_H5   = OUTPUT_DIR / "model.h5"
CLASS_PKL  = OUTPUT_DIR / "class_indices.pkl"

IMG_SIZE   = (224, 224)
TOP_K      = 5
ANNOTATE   = True  # set False if you don't want annotated images
ANNOTATED_DIR = OUTPUT_DIR / "annotated"

# ----------------------------
# Lazy globals (loaded once)
# ----------------------------
_MODEL = None
_IDX_TO_CLASS: Dict[int, str] = {}
_ORDERED_CLASSES: List[str] = []

# ----------------------------
# Helpers
# ----------------------------
def _ensure_artifacts():
    if not MODEL_H5.exists():
        raise FileNotFoundError(f"model.h5 not found at: {MODEL_H5}")
    if not CLASS_PKL.exists():
        raise FileNotFoundError(f"class_indices.pkl not found at: {CLASS_PKL}")

def _load_model_and_classes():
    global _MODEL, _IDX_TO_CLASS, _ORDERED_CLASSES
    if _MODEL is None:
        print("[INFO] Loading model…")
        _MODEL = load_model(str(MODEL_H5))

    if not _IDX_TO_CLASS:
        import pickle
        with open(CLASS_PKL, "rb") as f:
            class_indices = pickle.load(f)  # {'cardboard':0, ...}
        _IDX_TO_CLASS = {v: k for k, v in class_indices.items()}
        _ORDERED_CLASSES = [_IDX_TO_CLASS[i] for i in range(len(_IDX_TO_CLASS))]
        print("[INFO] Classes:", _ORDERED_CLASSES)

def _list_images(path: Path) -> List[Path]:
    exts = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp"}
    if path.is_file():
        return [path] if path.suffix.lower() in exts else []
    files = [p for p in path.rglob("*") if p.suffix.lower() in exts]
    return sorted(files)

def _load_tensor(img_path: Path) -> np.ndarray:
    img = Image.open(img_path).convert("RGB").resize(IMG_SIZE)
    arr = np.array(img).astype(np.float32)
    arr = preprocess_input(arr)
    return np.expand_dims(arr, axis=0)  # (1, H, W, 3)

def _annotate(image_path: Path, label_text: str, out_path: Path):
    img = Image.open(image_path).convert("RGB")
    draw = ImageDraw.Draw(img)
    try:
        font = ImageFont.truetype("arial.ttf", 24)
    except:
        font = ImageFont.load_default()

    margin = 8
    # get text bbox (Pillow 8+)
    try:
        bbox = draw.textbbox((0,0), label_text, font=font)
        text_w, text_h = bbox[2]-bbox[0], bbox[3]-bbox[1]
    except:
        text_w = int(draw.textlength(label_text, font=font))
        text_h = 24

    box_w, box_h = text_w + 2*margin, text_h + 2*margin
    draw.rectangle([(10, 10), (10+box_w, 10+box_h)], fill=(0,0,0,180))
    draw.text((10+margin, 10+margin), label_text, font=font, fill=(255,255,255))

    out_path.parent.mkdir(parents=True, exist_ok=True)
    img.save(out_path)

def _topk_from_probs(probs: np.ndarray, k: int) -> Tuple[List[str], List[float]]:
    idx = np.argsort(probs)[::-1][:k]
    classes = [_IDX_TO_CLASS[int(i)] for i in idx]
    scores  = [float(probs[int(i)]) for i in idx]
    return classes, scores

def _predict_tensor(batch: np.ndarray) -> np.ndarray:
    """Returns probabilities of shape (N, C)."""
    out = _MODEL.predict(batch, verbose=0)
    # Model already ends with softmax; if not, uncomment:
    # out = tf.nn.softmax(out, axis=-1).numpy()
    return out

def _df_preview_bar(df_row, k=5):
    """Inline bar chart for the top-k of a single prediction row (Jupyter visualization)."""
    labels = [df_row[f"top{i+1}_class"] for i in range(k) if f"top{i+1}_class" in df_row]
    scores = [df_row[f"top{i+1}_p"] for i in range(k) if f"top{i+1}_p" in df_row]
    if not labels: 
        return
    plt.figure(figsize=(6, 3))
    plt.barh(labels[::-1], scores[::-1])
    plt.xlim(0, 1.0)
    plt.xlabel("Confidence")
    plt.title("Top-K Prediction")
    plt.tight_layout()
    plt.show()

def _save_outputs(df: pd.DataFrame, stem: str):
    json_path = OUTPUT_DIR / f"{stem}.json"
    csv_path  = OUTPUT_DIR / f"{stem}.csv"
    df.to_json(json_path, orient="records", indent=2)
    df.to_csv(csv_path, index=False, encoding="utf-8")
    print(f"[INFO] Saved → {json_path}")
    print(f"[INFO] Saved → {csv_path}")

# ----------------------------
# Public notebook API
# ----------------------------
def predict_image(image_path: str, topk: int = TOP_K, annotate: bool = ANNOTATE) -> pd.DataFrame:
    """
    Predict a single image and show inline preview + top-k bar.
    Returns a one-row DataFrame.
    """
    _ensure_artifacts()
    _load_model_and_classes()

    p = Path(image_path)
    if not p.exists():
        raise FileNotFoundError(f"Image not found: {p}")

    batch = _load_tensor(p)
    probs = _predict_tensor(batch)[0]  # (C,)

    k = max(1, min(topk, len(_ORDERED_CLASSES)))
    classes, scores = _topk_from_probs(probs, k)
    pred_class, pred_conf = classes[0], scores[0]

    row = {
        "file": str(p),
        "pred_class": pred_class,
        "confidence": round(pred_conf, 6),
    }
    for i, (c, s) in enumerate(zip(classes, scores), start=1):
        row[f"top{i}_class"] = c
        row[f"top{i}_p"] = round(s, 6)

    df = pd.DataFrame([row])

    # Show dataframe
    display(df)

    # Show top-k bar
    _df_preview_bar(df.iloc[0], k=k)

    # Save annotated image (optional)
    if annotate:
        label = f"{pred_class} ({pred_conf*100:.1f}%)"
        out_img = ANNOTATED_DIR / f"{p.stem}_pred.png"
        try:
            _annotate(p, label, out_img)
            from IPython.display import display as ipy_display
            ipy_display(Image.open(out_img).resize((min(512, Image.open(out_img).width), 
                                                    min(384, Image.open(out_img).height))))
            print(f"[INFO] Annotated → {out_img}")
        except Exception as e:
            print(f"[WARN] Could not annotate: {e}")

    # Save per-call outputs
    _save_outputs(df, "predictions_single")

    return df

def predict_folder(folder_path: str, topk: int = TOP_K, annotate: bool = ANNOTATE, batch_size: int = 16) -> pd.DataFrame:
    """
    Predict all images under a folder (recursively). 
    Shows a small preview table + saves predictions.json/csv.
    """
    _ensure_artifacts()
    _load_model_and_classes()

    root = Path(folder_path)
    if not root.exists():
        raise FileNotFoundError(f"Folder not found: {root}")

    files = _list_images(root)
    if not files:
        raise FileNotFoundError(f"No images under: {root}")

    results = []
    k = max(1, min(topk, len(_ORDERED_CLASSES)))

    # Batch inference for speed
    for i in range(0, len(files), batch_size):
        batch_files = files[i:i+batch_size]
        batch = np.vstack([_load_tensor(p) for p in batch_files])
        probs_batch = _predict_tensor(batch)  # (N, C)

        for p, probs in zip(batch_files, probs_batch):
            classes, scores = _topk_from_probs(probs, k)
            pred_class, pred_conf = classes[0], scores[0]
            row = {
                "file": str(p),
                "pred_class": pred_class,
                "confidence": round(float(pred_conf), 6),
            }
            for j, (c, s) in enumerate(zip(classes, scores), start=1):
                row[f"top{j}_class"] = c
                row[f"top{j}_p"] = round(float(s), 6)
            results.append(row)

            if annotate:
                try:
                    label = f"{pred_class} ({pred_conf*100:.1f}%)"
                    out_img = ANNOTATED_DIR / f"{p.stem}_pred.png"
                    _annotate(p, label, out_img)
                except Exception as e:
                    print(f"[WARN] Annotate failed for {p.name}: {e}")

    df = pd.DataFrame(results).sort_values("confidence", ascending=False).reset_index(drop=True)
    # Preview top 10
    display(df.head(10))

    # Save outputs
    _save_outputs(df, "predictions")

    # Quick class distribution bar
    plt.figure(figsize=(6,3))
    df["pred_class"].value_counts().plot(kind="bar")
    plt.ylabel("# Images")
    plt.title("Predicted Class Distribution")
    plt.tight_layout()
    plt.show()

    if annotate:
        print(f"[INFO] Annotated images → {ANNOTATED_DIR}")

    return df

# ----------------------------
# Convenience: if you just run this cell,
# it will try a quick demo by picking one image
# from your dataset-resized\plastic folder (if present).
# ----------------------------
try:
    sample_dir = DATA_ROOT / "plastic"
    sample_files = _list_images(sample_dir) if sample_dir.exists() else []
    if sample_files:
        print("[DEMO] Running one quick prediction from 'plastic' folder to show output…")
        _ = predict_image(str(sample_files[0]), topk=3, annotate=False)
    else:
        print("[DEMO] No sample found in:", sample_dir)
except Exception as e:
    print("[DEMO] Skipped demo:", e)
```
